```python
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

#Wykres liniowy

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

#Tworzenie wielu wykresów

```python
plt.subplot(121)  #XYZ x dł osi x, yilość słupków, z nr wykresu
plt.plot([0,1,2] , [3,2,5])
plt.subplot(122)
plt.plot([0, 1,2],[3,1,4], 'r--')
```

```python
fig = plt.figure(figsize=(12,8)) # możemy sterować rozmiarem tworzonych wykresów
plt.subplot(131)  #XYZ x dł osi x, yilość słupków, z nr wykresu
plt.plot([0,1,2] , [3,2,5])
plt.subplot(132)
plt.plot([0, 1,2],[3,1,4], 'r--') # czerwona przerywana linia
plt.subplot(133)
plt.plot([0, 1,2],[3,1,4], 'ro')  # czerwone punkty

fig.show()
```

#Wykresy słupkowe

```python
plt.bar(x=[0,1,2,3], height=[1.2,2.3, 1.2, 5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2,2.3, 1.2, 5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2,2.3, 1.2, 5.2])  #wertykalny
```

#Wykresy punktowa

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300) * 50
plt.scatter(x=x1, y=x2, s=x3)  # s - rozmiar kropki
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)  # alpha - przezroczystość
```

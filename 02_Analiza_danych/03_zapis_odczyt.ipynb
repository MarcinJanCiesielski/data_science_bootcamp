#Import bibliotek

```python
import numpy as np
import pandas as pd
```

#Za≈Çadowanie danych

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df.head()
```

#Zapis/odczyt danych z pliku CSV

```python
df.to_csv('amzn.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2022)]
df_nov
```

```python
df_nov.to_csv('amzn_nov.csv')
```

```python
new_df = pd.read_csv('amzn_nov.csv')
new_df
```

```python
df_nov.to_excel('amzn_nov.xlsx')
```

# Maski logiczne

```python
a = np.arange(start=-10, stop=10, step=0.5)
```

```python
a = a.reshape(10, -1)
a
```

```python
a > 0
```

```python
a[a > 0]
```

```python
np.bitwise_and(a > -5, a < 5)
```

```python
a[np.bitwise_and(a > -5, a < 5)] # bitowa koniunkcja (boot)
```

```python
np.bitwise_or(a < -5, a > 5) # bitowa alteratywa(bool)
```
